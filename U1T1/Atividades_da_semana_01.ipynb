Knowing this information and knowing the probabilities listed in the image above it is possible to calculate all the probabilites like this: For both ocurrances to happen we have to multiply the probabilities for each case.

H-H = p1(H) x p2(H) = 0,5 x 0,5 = 0,25 --> The probability of coin 1 to show Head AND coin 2 to show head.

T_H = (1-p1(H)) x p2(H) --> The probability of coin 1 to show Tail AND coin 2 to show Head.

T_T = (1-p1(H)) x (1-p2(H)) --> The probability of coin 1 to show Tail AND coin 2 to show Tail.

### Exercise 2.3:
You're given the string: "OCZ XJMMZXO VINRZM". Each letter follows a stochastic Markov process with the rules expressed by this table. Follow the process for three steps and reconstruct the correct answer. (Note, this is a Caesar cipher with shift 7 applied three times, because the Caesar cipher is a Markov process).

```python
def decifrar_cesar(texto, deslocamento):
    resultado = ""

    for char in texto:
        # Verifica se é uma letra maiúscula
        if char.isupper():
            # Calcula o novo caractere com base no deslocamento
            novo_char = chr((ord(char) - deslocamento - 65) % 26 + 65)
            resultado += novo_char
        # Verifica se é uma letra minúscula
        elif char.islower():
            novo_char = chr((ord(char) - deslocamento - 97) % 26 + 97)
            resultado += novo_char
        else:
            # Não altera caracteres que não são letras
            resultado += char

    return resultado

origin = "OCZ XJMMZXO VINRZM"

deslocamento = 7
texto_decriptado = decifrar_cesar(origin, deslocamento)
print("Primeira Descriptografia/deslocamento:", texto_decriptado)
texto_decriptado1 = decifrar_cesar(texto_decriptado, deslocamento)
print("Segunda Descriptografia/deslocamento:", texto_decriptado1)
texto_decriptado2 = decifrar_cesar(texto_decriptado1, deslocamento)
print("Terceira Descriptografia/deslocamento:", texto_decriptado2)
```

### Exercise 2.4:
Draw the probability mass function and the cumulative distribution of the following outcome probabilities:


image.png

```python
import numpy as np
import matplotlib.pyplot as plt

# Amostras
outcomes = np.array([1, 2, 3, 4, 5, 6, 7, 8])
probabilities = np.array([0.1, 0.15, 0.2, 0.21, 0.17, 0.09, 0.06, 0.02])

# Este código verifica se a soma das probabilidades para cada caso é igual a 1
if not np.isclose(np.sum(probabilities), 1):
    raise ValueError("A soma dos valores de probabilidade devem ser 1.")

# Plotting the Probability Mass Function
plt.stem(outcomes, probabilities)
plt.title("Probability Mass Function")
plt.xlabel("Resultados")
plt.ylabel("Probabilidade")
plt.xticks(outcomes)
plt.grid()
plt.show()

prob_accumulative = np.zeros(len(probabilities))

# Calcula a soma acumulativa
for i in range(len(probabilities)):
    if i == 0:
        prob_accumulative[i] = probabilities[i]
    else:
        prob_accumulative[i] = prob_accumulative[i - 1] + probabilities[i]

print(prob_accumulative)

plt.stem(outcomes, prob_accumulative)
plt.title("Função de distribuição acumulativa")
plt.xlabel("Resultados")
plt.ylabel("Probabilidade acumulativa")
plt.xticks(outcomes)
plt.grid()
plt.show()
```

### Exercise 2.5:
How many bits do we need to independently encode v1 and v2 from this [data](https://www.networkatlas.eu/exercises/2/5/data.txt)? How much would we save in encoding v1 if we knew v2?



Para calcular a quantidade de bits necessários para codificar V1 é necessário saber a probabilidade de cada resultado para V1, ou seja: p*-log2(p), para todo p, onde p representa a probabilidade de ocorrência de cada elemento único de V1. Então, haverá p(a), p(b), p(c) até p(f).
Com os valores de p para cada elemento, soma todas informações de ocorrência - como definido no livro - fazendo o somatório de p * - log(p). O resultado será a entropia de v1.

O mesmo pode e deve ser feito com v2, assim poderá ser calculada a porcentagem de ocorrências simultâneas. A definição de **informação simultânea** de acordo com a documentação do sktlearn.metrics:

image.png

```python
###                   A maioria do código é do exemplo do github disponível no site de execícios.

import pandas as pd
from scipy.stats import entropy
from sklearn.metrics import mutual_info_score
# Load the data
df = pd.read_csv("https://www.networkatlas.eu/exercises/2/5/data.txt", sep = "\t")
##print(df)
# The number of bits we need to encode v1 and v2 is their entropy. To calculate it, we need the probability of each possible outcome:
v1_ps = (df.groupby(by = "v1").size() / df.shape[0]).values
v2_ps = (df.groupby(by = "v2").size() / df.shape[0]).values

##print(v1_ps)
##print(v2_ps)

v1_info = entropy(v1_ps)
v2_info = entropy(v2_ps)

print("Probabilidade de ocorrências para cada elemento único de v1:", v1_info)
print("Probabilidade de ocorrências para cada elemento único de v2:", v2_info)

# The bits v2 gives us about v1 is their mutual information.
mi = mutual_info_score(df["v1"], df["v2"])
# Thus we save ~ 0.79 bits out of ~ 1.75 or ~ 45%
#print("Knowing v2 gives us %1.2f%% of v1" % (100 * mi / v1_info))
print("Saber das elementos de v2 nos dá %1.2f%% de v1" % (100 * mi / v1_info))
```

Draw a correlation network for these vectors, by only drawing edges with positive weights, ignoring self loops.

Uma rede de correlações é uma representação gráfica de elementos semelhantes entre diferentes fontes de dados. Temos oito vetores e precisamos representar as correlações entre eles.
Não achei no livro algo explicando ou falando qual a fórmula ou forma de cálcular esta correlação, porém, pesquisando vi que a documentação do pandas.DataFrama.corr() usado no código exemplo usa a fórmula de Pearson como padrão para o cálculo. Com a ajuda do ChatGPT pude ver que a fórmula é definida assim:



image.png

n é o número de pares de dados.

∑𝑥𝑦 é a soma dos produtos dos pares.

∑𝑥 é a soma de todos os elementos do vetor X.

∑𝑦 é a soma de todos os elementos do vetor Y.

O cálculo das correlações é feito com pares de vetores, ou seja, é calculada a correlação de A com B, A com C, A com D, até todos os pares serem calculados.
Para cada par de vetor, é calculada a correlação entre os elementos de mesma posição, ou seja: Correlação de A[0] com B[0], A[1] com B[1] até todos os pares de mesma posição dos vetores serem calculados.

A soma das correlações dos pares de elementos é a correlação dos vetores, após o cáculo das correlações pode ser feita uma matriz de adjecências para melhor visualizar a correlação entre vetores como é feito no código abaixo.

A última parte do código faz a remoção de valores de correlação negativos e a correlação entre o mesmo vetor para refpresentação gráfica.

```python
import pandas as pd

import networkx as nx
import matplotlib.pyplot as plt
# Read the data
df = pd.read_csv("https://www.networkatlas.eu/exercises/3/4/data.txt", sep = "\t")
# Calculate correlations
df = df.corr()

## Exibir a matriz de correlação
print(df)

## Retirando valores de correlações negativas e correlações entre o mesmo vetor para representação gráfica.
# Take only positive correlations
df = df > 0
df = df.unstack().reset_index()
df.columns = ("source", "target", "edge")
df = df[df["edge"]]
# Remove self loops
df = df[df["source"] != df["target"]]
# Make networkx object
G = nx.from_pandas_edgelist(df)
# Draw
nx.draw(G, with_labels = True)
plt.savefig("correlation_network.png")
plt.show()
```

This network (https://www.networkatlas.eu/exercises/4/1/data.txt) is bipartite. Identify the nodes in either type and find the nodes, in either type, with the most neighbors.

No link acima há uma tabela com duas colunas, cada elementos da coluna representa um nó e cada linha representa uma ligação entre os nós da respectiva posição.

Por ser uma rede bipartida os elementos da coluna esquerda não podem aparecer na coluna direita e vice-versa; os elementos da mesma coluna não podem ser ligados entre si. Porém, pode haver várias ligações entre elementos de colunas diferentes e pode haver caminho entre elementos de mesma coluna através de ligações com a outra coluna.

A quantidade de vizinhos é definida pela quantidade de elementos que se conectam com um nó. A nó com mais vizinhos em uma rede bipartida é o nó que faz mais ligações com nós da outra coluna.

Não consegui rodar o código e tá ruim de contar na mão, porém, para o grupo da primeira coluna o nó com mais ligações diretas é o nó 131 com 4 vizinhos e para o segundo grupo o nó com mais vizinhos é o nó 2 com 59 ligações.

```python
## NÃO CONSEGUI RODAR O CÓDIGO TUTORIAL

import networkx as nx

# Load the data
G = nx.read_edgelist("https://www.networkatlas.eu/exercises/4/1/data.txt")
# Divide the nodes in the two bipartite sets
nodes = nx.algorithms.bipartite.basic.sets(G)
# Find the node in type 0 with the most neighbors
node_neighbors = {n: len(set(G.neighbors(n))) for n in nodes[0]}
maxnode = max(node_neighbors, key = node_neighbors.get)
print(maxnode, node_neighbors[maxnode])
# Find the node in type 1 with the most neighbors
node_neighbors = {n: len(set(G.neighbors(n))) for n in nodes[1]}
maxnode = max(node_neighbors, key = node_neighbors.get)
print(maxnode, node_neighbors[maxnode])
```
